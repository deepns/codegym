```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Load model
model_name = "microsoft/Phi-3-mini-4k-instruct"
model = AutoModelForCausalLM.from_pretrained(model_name,
                                             device_map="cuda",
                                             torch_dtype="auto",
                                             trust_remote_code=False)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Starting with some sample params for the pipeline as
# given in the book
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=500,
    do_sample=False
)

# The model used above doesn't seem to give factual answers right
# When asked the following question, it answered West Indies
messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Who won the cricket world cup in 2011?"},
]

# Generate output
response = generator(messages)
print(response[0]["generated_text"])
```
